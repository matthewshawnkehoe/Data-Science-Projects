A data science project consists of several phases. In this module, you'll be introduced to the first phase: *exploratory data analysis* (EDA). EDA is first for a good reason—every subsequent phase of any data science project depends on it for success. As a data scientist candidate, it's essential to have a strong command of EDA before you move on to more advanced topics.

As the name suggests, exploratory data analysis is the process of exploring data to discover insights, identify patterns, establish relationships and trends, and test assumptions. The best ally that a data scientist can have is the data itself. However, getting to know the data takes patience, openness, and attention. EDA is the process of letting the data speak for itself.

When working with a dataset, you might have an objective like the following:

* Predicting next month's housing prices
* Predicting customer churn
* Predicting a plant's electricity consumption for tomorrow
* Building a mobile app that can tag people in a photo
* Implementing a recommendation system for a video-streaming website

All of the tasks listed above can be accomplished by one or more data scientists if relevant, clean datasets are provided. However, just like money, clean datasets do not grow on trees! Usually, this data comes from raw, messy formats. 

## Features and targets

In data science, measurable properties of datasets are referred to as *features* and *targets*. Loosely, the distinction is equivalent to that of *independent* and *dependent* variables in classical statistics; however, the ends and means are different here. As discussed in the beginning of this program, data science is different from statistics. 

The target variable of a dataset is the particular feature that you want to understand better. Often, you want to predict this variable. In supervised learning, you will use data to find relationships between other features and the target. 

In order to get the most out of the data, you need to eliminate irrelevant and potentially misleading parts, while also transforming them into a format that you can work with easily. Editing is a crucial aspect of EDA. You need to improve the quality of the data—and improving data quality usually means *data cleaning*. This process is also referred to as *data munging* or *data wrangling*. 

As you clean, you'll begin to notice the useful features that are associated with your target. This process, known as *data exploration*, analyzes the relationships between features as well as between the features and the target. After you have cleaned and explored your data, identifying and creating useful features is called *feature engineering*.

EDA, then, is the combination of data cleaning, data exploration, and feature engineering. Now that you understand all the components, review what you've learned by checking out the EDA process overview below. 


# What is exploratory data analysis?

Exploratory data analysis is the building block of the data science pipeline. Before moving into the modeling phase, you need to understand the relationships in your data. And the more you understand your data, the more you can successfully identify the most useful features that will help your models achieve the best performance. 

You can conceptualize EDA as an iterative process consisting of three steps. The figure below depicts these steps and how the iteration occurs:

1. **Data cleaning:** This is the first step of EDA. In this step, you will take a raw dataset and eliminate any problems that would prevent you from further analysis. 

2. **Data exploration:** The second step is exploring the data to discover relationships and features. You will explore the data using both statistics and data visualization. If you identify new problems in the data during exploration, you will go back to the cleaning phase and eliminate these identified problems.

3. **Feature engineering:** The final step of EDA is the feature engineering, where you select the most useful features or create new features from the existing ones. This phase leads you to the next phases of the data science pipeline. However, if you detect any problems with the data in this phase, you will return to the data-cleaning step. And if you transform existing features or create new ones, you will need to go back to the data-exploration phase to investigate them.
