## Statistical models
Models are ubiquitous. You rely on models every time that you [check the weather report](https://www.weather.gov/mdl/StatisticalModeling_home), [choose a movie recommended by Netflix](https://www.theatlantic.com/technology/archive/2014/01/how-netflix-reverse-engineered-hollywood/282679/), or [get excited about your favorite presidential candidate's electoral potential](https://fivethirtyeight.com/features/a-users-guide-to-fivethirtyeights-2016-general-election-forecast/).

In data science, the word *models* usually refers to *statistical models*. Statistical models are significantly different from *mathematical models*. In mathematical modeling, you formulate a phenomenon in a very deterministic sense, and there's no room for uncertainty. But in statistical modeling, you incorporate uncertainty into the model. For example, think about tossing a fair coin. Can you write down an exact formula to determine whether heads or tails will come in a coin flip? Actually, containing the exact process in an exact formula is very complicated! But you can formulate the problem in a statistical model very easily. If you repeat the process several times, you'll see that half the time it turns out to be tails, and half the time it is heads. So, your statistical model would be as follows: with 0.5 probability, tails would occur, and with 0.5 probability, heads would occur.

Statistical models are, in essence, simple representations of the complex reality. In that sense, they are inherently wrong formulations of the phenomenon that they try to capture. But they are extremely useful for many of the tasks that you are dealing with. To understand this point better, take a look at an example:

If you've ever navigated public transit, you've worked with a model. Possibly the most famous public transit map in the world is for the London Underground (commonly known as "the Tube"):

image.png

This map represents all of the train lines in the London Underground in terms of their stops and intersections with one another. It's a simplified representation of the real train lines. For example, it does not include any information on the relative distances between the stops or the lines' actual geographical positions. When it was first proposed, this model of the London Underground was revolutionary: it excludes information that is irrelevant to the needs of Tube travelers—and by doing so, it becomes much easier to use.

Statistical models operate in the same way: they represent complex information in a simplified way to serve a particular goal. To be more exact, a statistical model is a simplified mathematical representation of the data scientist's best guess about the underlying processes that created the data.

That definition has a lot of moving parts. Below, go through it piece by piece.

### Simplified

When creating a model, data scientists prioritize *dense features*: elements with information that explains a large amount of variance in the outcome of interest. By excluding less informative features, the model does not perfectly represent the data, but its simplified form is easier to understand and work with. After all, reality is too complex to be modeled perfectly.

For the London Tube Map, the goal was to guide travelers through the system. The position of each stop on a line and the intersection of lines are dense features for travelers. On the other hand, the relative time between adjacent stops is less relevant, so it's excluded from the map.

### Mathematical representation

A statistical model is a mathematical formula or series of formulas. It can also be represented using words or programming statements, but these are just glosses on the math. Here, the London Tube Map isn't a great example. Because it's a conceptual model rather than a mathematical one, it's represented visually rather than mathematically.

### The data scientist's best guess

Generally speaking, the model that you propose and test represents your best understanding of the data at that time. That means it's your best guess about the underlying processes that led to the outcomes that you observe. The model design is informed by prior knowledge about the topic area, visual inspection of the data, and the outcomes of fitting previous models to the data.

The London Tube Map has actually been [improved several times since its original version](https://londonist.com/2016/05/the-history-of-the-tube-map) as its designers learned more about how people were processing and understanding the information on the map.

### Underlying processes

A good model gives some insight into how your variable of interest came to have the values it possesses. This aspect of modeling is particularly useful when you are interested in predicting future outcomes. If you have a reasonable approximation of the underlying processes that created the outcomes that you have in the data, you can use the model to predict how those underlying processes would respond to new data and what outcomes would result.

By showing all the Tube lines and how they intersect, the London Tube Map not only shows travelers where they need to go, but also gives them a general idea of what will happen to get them there. Imagine a subway navigation system without any underlying process information—you would give your starting point and destination, and you'd be given a list of stops and exchanges to use—but with no context about where they are in relation to each other.

It is the goal of every data scientist to create statistical models that are as efficient and useful as the London Tube Map! With the help of a good statistical model, even a nonstatistician can understand a complex dataset. A statistical model can be beautiful; by paring away the nonessential aspects of a dataset, you get a tiny glimpse into the workings of the universe.
